### EVEN AND ODD

```python
n=int(input())
m=int(input())
for i in range(n,m):
    if(i%2==0):
        print("even",i)
else:
    print("odd",i)
```

# factorial

```python
a=int(input())
i=1
f=1
for i in range(1,a+1):
    f=f*i
    i=i+1
    print(f)
    
```
